## IQR Calculating and Visulize

```python
import matplotlib.pyplot as plt
ages = [22, 25, 30, 45, 40, 23, 35, 40, 38, 26]

sorted_age = sorted(ages)
ages.sort()
n = len(ages)
mid = n // 2

if  n % 2 == 0:
    q1 = ( ages[mid // 2 -1] + ages[mid // 2] ) / 2
else:
    q1 = ages[mid // 2 ]
    
if n % 2 == 0:
    q3 = (ages[mid + mid// 2] + ages[mid + mid // 2+1 ] ) / 2
else:
    q3 = ages[mid + mid //2 +1]

print(q1)
print(q3)
IQR = q3 - q1
print(IQR)

# visual 


plt.boxplot(ages,vert=False,patch_artist=True,boxprops=dict(facecolor="lightblue"))

plt.axvline(q1, color='blue', linestyle='--', label=f"Q1: {q1}")
plt.axvline(q3, color='green', linestyle='--', label=f"Q3: {q3}")

plt.xlabel("Ages")
plt.title("Box Plot of Ages with Q1, Q3, and IQR")
plt.legend()

plt.show()
print("Interquartile-Range",IQR)
```

## Mean, Median , Mode

```python
data = data = [22, 24, 23, 21, 25, 26, 24, 23, 22, 21, 20, 22, 24, 25, 26, 
        28, 30, 29, 27, 26, 25, 24, 23, 22, 21, 22, 23, 24, 25, 26]

data.sort()
# mean
mean = sum(data) / len(data)
n = len(data)

#Median
if n % 2 == 0:
    median = (data[n//2 -1] + data[n // 2]) / 2
else:
    median = data[n//2]
    
# Mode
freq = {}
for i in data:
    if i in freq:
        freq[i] +=1
    else:
        freq[i] = 1
        
max_rep = max(freq.values())

res = [k for k,v in freq.items() if v == max_rep]

print(mean)
print(median)
print(res[0])
```

```python
## Visulaization 

import matplotlib.pyplot as plt

plt.hist(data)                    #histogram ---> Understanding Data Distribution 


plt.axvline(mean,color="blue",linestyle="--",label=f"Mean = {mean}")
plt.axvline(res[0],color="g",linestyle="--",label=f"mode = {res[0]}")
plt.axvline(median,color="red",linestyle="--",label=f"median = {median}")
plt.legend()

plt.show()
```

## STD AND VARIENCE


```python
data = [22, 24, 23, 21, 25, 26, 24, 23, 22, 21, 20, 22, 24, 25, 26, 
        28, 30, 29, 27, 26, 25, 24, 23, 22, 21, 22, 23, 24, 25, 26]

mean = sum(data) / len(data)

sq_diff = [(x-mean) ** 2 for x in data]
variance = sum(sq_diff) / len(data)

std = variance ** 0.5


# A scatter plot is used to visualize the relationship or distribution of individual data points along two axes,

plt.scatter(data, [0]*len(data), color='blue', label='Data Points')

# Plot Mean line
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean}')

# Plot Standard Deviation Range lines
plt.axvline(mean + std, color='green', linestyle=':', label=f'Mean + 1 Std Dev')
plt.axvline(mean - std, color='green', linestyle=':', label=f'Mean - 1 Std Dev')

plt.title("Variance & Standard Deviation Visualization")
plt.legend()

# Show the plot
plt.show()

# Print Variance and Standard Deviation
print(f"Variance: {variance}")
print(f"Standard Deviation: {std}")
```

## Pearson Correlation

```python
x = [10, 20, 30, 40, 50]
y = [15, 25, 35, 45, 55]

mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)

# Calculate covariance
cov = 0
for i in range(len(x)):
    cov += (x[i] - mean_x) * (y[i] - mean_y)

cov_res = cov / (len(x) - 1)  # Use (n - 1) for sample covariance

# Calculate standard deviations for x and y
deno_x = (sum((i - mean_x) ** 2 for i in x) / (len(x) - 1)) ** 0.5
deno_y = (sum((i - mean_y) ** 2 for i in y) / (len(y) - 1)) ** 0.5

# Calculate Pearson correlation
denominator = deno_x * deno_y
pear = cov_res / denominator

# A scatter plot is used to visualize the relationship or distribution of individual data points along two axes,

plt.scatter(x, y, color='blue')

# Adding titles and labels
plt.title(f"Pearson Correlation: {pear:.2f}")
plt.xlabel("X")
plt.ylabel("Y")

# Show plot
plt.show()

# Print Pearson Correlation Coefficient
print(f"Pearson Correlation Coefficient: {pear}")
```

```python
# A company is testing if a new training program improves employee productivity.
```

```python
# here we use   paired t-test       for before after scenario

import pandas as pd

data = {
     "Before": [75, 80, 85, 70, 65, 90, 78, 84, 76, 88, 92, 85, 79, 87, 72, 80, 77, 83, 89, 74],
    "After": [85, 88, 92, 78, 72, 95, 84, 89, 82, 91, 98, 90, 86, 92, 79, 85, 82, 90, 94, 81],
}

df = pd.DataFrame(data)
# now lets do T-test

from scipy.stats import ttest_rel

t_stat,p_value = ttest_rel(df["Before"],df["After"])
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")

if p_value <= 0.05:
    print("reject")
else:
    print("fail to reject")
```

## Chi-Squre Test 

```python
from scipy.stats import chisquare
```

```python
# 1) A company wants to check if customer preferences for four product categories are evenly distributed
# They surveyed customers, and the observed counts are as follows:

#Category	Observed Count
#Electronics	40
#Clothing	35
#Groceries	25
#Furniture	20
```

```python
# Chi-Square Goodness-of-Fit Test

from scipy.stats import chisquare
observed = [40, 35, 25, 20]

total_customer = sum(observed)
expected = [total_customer / 4] * 4


# test 
chi_stat,p_value = chisquare(f_obs=observed,f_exp=expected)

print(f"Chi-Square Statistic: {chi_stat}, P-Value: {p_value}")


if p_value <= 0.05:
    print("reject")
else:
    print("fail to reject")
```

```python
# visualize

import matplotlib.pyplot as plt
categories = ["Electronics", "Clothing", "Groceries", "Furniture"]
plt.bar(categories, observed, label="Observed", alpha=0.7)
plt.bar(categories, expected, label="Expected", alpha=0.7)
plt.ylabel("Count")
plt.title("Observed vs Expected Customer Preferences")
plt.legend()
plt.show()
```

# ANOVA- Test

```python
# A company wants to determine if there is a significant difference in the average sales generated by three
#different advertising strategies: TV Ads, Online Ads, and Social Media Ads.
```

```python
# one way anova test            # atleast one group is different

import scipy.stats as stats

sales_tv = [50, 60, 55, 70]
sales_online = [65, 70, 60, 75]
sales_social = [45, 50, 40, 55]

f_stat,p_value = stats.f_oneway(sales_tv,sales_online,sales_social)

print(f"F-Statistic: {f_stat}, P-Value: {p_value}")

if p_value < 0.05:
    print("At least one group is significantly different.")
else:
    print("No significant difference between groups.")
```

```python
# visulize 

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = pd.DataFrame(
 {"Sales": sales_tv + sales_online + sales_social,
  "Ad Startegy" : ["Tv"]*4 + ["online"]*4 + ["Social"] *4
 }
)

sns.boxplot(x="Ad Startegy",y="Sales",data=data)
plt.title("Sales Performance Across Ad Strategies")
plt.show()
```

# Z-Test

```python
# A coffee shop claims that the average time to serve a customer is 5 minutes.
# You collected a sample of 40 service times with an average of 5.5 minutes and 
# a standard deviation of 1.5 minutes. Test if the service time is significantly
# different from the claim at a 5%  significance level.
```

```python
# one-sample z-Test 

import scipy.stats as stats
import numpy as np

sample_mean = 5.5
population_mean = 5
population_std = 1.5
sample_size = 40

z_stat = (sample_mean / population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1- stats.norm.cdf(abs(z_stat)))

print(f"Z-Statistic: {z_stat:.2f}")
print(f"P-Value: {p_value:.4f}")

# Create a range of Z values
z_values = np.linspace(-4, 4, 1000)
pdf_values = stats.norm.pdf(z_values)

if p_value < 0.05:
    print("Reject the null hypothesis: The service time is significantly different.")
else:
    print("Fail to reject the null hypothesis: The service time is not significantly different.")
```

```python
# visulize 

import matplotlib.pyplot as plt
plt.plot(z_values, pdf_values, label="Standard Normal Distribution", color="blue")

# Highlight the rejection region
plt.fill_between(z_values, pdf_values, where=(z_values > abs(z_stat)), color='red', alpha=0.5, label="Rejection Region")

# Mark the Z-statistic on the plot
plt.axvline(x=z_stat, color='green', linestyle='--', label=f"Z-Statistic: {z_stat:.2f}")

# Labels and title
plt.title("Z-Test Visualization")
plt.xlabel("Z-Value")
plt.ylabel("Probability Density")
plt.legend()

plt.show()
```
