```python
data =[5, 12, 19, 26, 33]

min_d = min(data)
max_d = max(data)
Range = max_d - min_d

print("Range is :",Range)


import matplotlib.pyplot as plt

plt.plot(data, 'o', label='Data Points')
plt.axhline(min_d, color='r', linestyle='--', label='Min')
plt.axhline(max_d, color='g', linestyle='--', label='Max')
plt.legend()
plt.show()
```

```python
data =[ 8, 10, 10, 10, 12, 12, 14]

mean = sum(data) / len(data)

sq_dif = [(x- mean)**2 for x in data]

Varience = sum(sq_dif) / len(data)
print("Varience is : ",Varience)

std = Varience ** 0.5
print("Standard Deviation ",std)


import matplotlib.pyplot as plt
plt.plot(data,'o',label='data point')
plt.axhline(mean,color="red",linestyle="--",label="Mean")

plt.axhline(mean + std, color="green",linestyle="--",label="Mean + std")
plt.axhline(mean - std, color="green",linestyle="--",label="Mean - std")
plt.legend()
plt.show()
```

```python
data = [4, 7, 9, 12, 15, 18, 21, 24, 30]
data.sort()

n = len(data)
mid = n //2

if n % 2 ==0:
    q1 = (data[mid // 2-1] + data[mid //2] )/ 2
else:
    q1 = data[mid // 2]
    
    
if n % 2 == 0:
    q3 = (data[mid + mid // 2-1] + data[mid + mid //2]) / 2
else:
    q3 = data[mid + mid //2+1]

iqr = q3 - q1
print("Q1 is ",q1)
print("Q3is ",q3)
print("IQR is :" ,iqr)

    
import matplotlib.pyplot as plt
plt.boxplot(data,vert=False)
plt.title('Box Plot of Data')
plt.xlabel('Value')
plt.axvline(q1, color='r', linestyle='--', label='Q1')
plt.axvline(q3, color='g', linestyle='--', label="Q3")
plt.show()
```

```python

x = [2, 4, 6, 8, 10]
y = [50, 55, 60, 65, 70]

mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)

cov = 0
for i in range(len(x)):
    cov += (x[i]- mean_x )* (y[i] - mean_y)

covarience = cov / len(x)

print("Covarience is : ",covarience)



import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
```

```python
## Correlation

#1 Pearson method

t = [25, 30, 35, 40, 45] 
I = [200, 250, 300, 350, 400]

mean_t = sum(t) / len(t)
mean_i = sum(I) / len(I)

cov = 0
for i in range(len(t)):
    cov += (t[i] - mean_t) * (I[i] - mean_i)
        
covarience = cov / len(t)


dev_t = (sum((x- mean_t)**2 for x in t) / len(t)) ** 0.5
dev_I =(sum((y - mean_i)**2 for y in I) / len(t)) ** 0.5

deno = dev_t * dev_I

pearson = (covarience / deno)

print("Pearson Correlation :",pearson)



import matplotlib.pyplot as plt

plt.scatter(t,I)
plt.xlabel("T")
plt.ylabel("I")
plt.show()
```

```python
# Spearman Rank Correlation

from scipy.stats import spearmanr

x = [2, 4, 6, 8, 10]
y = [50, 55, 60, 70, 80]

rho,p_value = spearmanr(x,y)

print("Spearman Correlation:", rho)
print("P-value:", p_value)


import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Probability Distribution 

```python
#1  uniform distribution

import matplotlib.pyplot as plt
import numpy as np

a = 0  #min
b = 10 # max
size = 100

data = np.random.uniform(a,b,size)
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.hist(data,bins=10,edgecolor='black')
```

```python
# 2 Normal Ditribution 

a = 0 #mean
b = 1 #std
size = 100
data = np.random.normal(0,1,100)

plt.hist(data)
plt.show()
```

```python
# 3 Binomial Ditribution

n = 10  # no of trails
p = 0.5  # probability of success
size = 100

data = np.random.binomial(n,p,size)

plt.hist(data,bins=range(0,n+2),edgecolor="black")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4 Poisson Ditribution

lam = 3 # avg no of event
size = 100

data = np.random.poisson(lam,size)

plt.hist(data,bins=20)
plt.xlabel("Number of Events")
plt.ylabel("Frequency")
plt.show()
```

```python
# 5 Bernoulli Distribution

p = 0.5
size = 100

data = np.random.binomial(1,p,size)

plt.hist(data)

plt.xlabel("Outcome (0 = Failure, 1 = Success)")
plt.ylabel("Frequency")
plt.show()

```

# Hypothesis Test

```python
# T-test   

from scipy import stats

data = [172, 160, 165, 175, 180, 169, 168, 171, 173, 179]

p_mean = sum(data) / len(data)


t_stat,p_value = stats.ttest_1samp(data,p_mean)

print("T-static :",t_stat)
print("P value is :",p_value)
```
